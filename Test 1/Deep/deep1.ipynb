```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint

# Load your dataset
dataset = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\CKDu_processed.csv')
data = dataset.iloc[:, 1:33]  # Select columns 1 to 32 (32 features)
target = dataset.iloc[:, 0]

# Standardize your data
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(data)

# Assuming 'target' is your array of integer labels
from tensorflow.keras.utils import to_categorical
new_target = to_categorical(target)

# Split your data into training and testing sets
train_data, test_data, train_target, test_target = train_test_split(
    standard_scaled_data, new_target, test_size=0.2)

# Create your neural network model
model = Sequential()
model.add(Dense(512, input_dim=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='relu'))
model.add(Dense(6, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Create a ModelCheckpoint callback to save the best model
checkpoint = ModelCheckpoint(
    'best_model.h5',  # Specify the filename for the best model
    monitor='val_loss',
    save_best_only=True,
    mode='auto'
)

# Train your model and use the checkpoint
history = model.fit(train_data, train_target, epochs=1000,
                    validation_data=(test_data, test_target),
                    callbacks=[checkpoint])

# The best model based on validation loss will be saved as 'best_model.h5'
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout


# Load the CSV file
df = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\NEandE2.csv')

# Select columns for the 3D plot
data = df[['V', 'Hg', 'Ba']]
target = df['Target']

# Standardize the input features
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(data)

# Convert target labels to one-hot encoded format
new_target = to_categorical(target)

# Split the data into training and testing sets
train_data, test_data, train_target, test_target = train_test_split(standard_scaled_data, new_target, test_size=0.2)

model = Sequential()

# Add layers to the model
model.add(Dense(416, input_dim=3, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display the model summary
model.summary()
# Train the model
history = model.fit(train_data, train_target, epochs=500, validation_data=(test_data, test_target))

from matplotlib import pyplot as plt

# Plot loss
plt.figure(figsize=(10, 6))
plt.plot(model.history.history['loss'], label='Training Loss')
plt.plot(model.history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()

# Plot accuracy
plt.figure(figsize=(10, 6))
plt.plot(model.history.history['accuracy'], label='Training Accuracy')
plt.plot(model.history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training and Validation Accuracy')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc

# Make predictions on the test set
predictions = model.predict(test_data)
predicted_labels = (predictions > 0.5).astype(int)  # Convert probabilities to binary labels

# Calculate accuracy
accuracy = accuracy_score(test_target, predicted_labels)
print(f"Accuracy: {accuracy:.2f}")

# Calculate precision, recall, and F1-score using "micro" averaging
precision_micro = precision_score(test_target, predicted_labels, average='micro')
recall_micro = recall_score(test_target, predicted_labels, average='micro')
f1_micro = f1_score(test_target, predicted_labels, average='micro')

print(f"Precision (micro): {precision_micro:.2f}")
print(f"Recall (micro): {recall_micro:.2f}")
print(f"F1-score (micro): {f1_micro:.2f}")

# Make predictions on the test set
predictions = model.predict(test_data)
fpr, tpr, _ = roc_curve(test_target[:, 1], predictions[:, 1])  # Assuming the second class (index 1) is the positive class
roc_auc = auc(fpr, tpr)

# Print ROC AUC value
print(f"ROC AUC: {roc_auc:.2f}")

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```
