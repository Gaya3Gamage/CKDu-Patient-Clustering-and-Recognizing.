```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Load data into a pandas DataFrame
df = pd.read_csv('C:/Users/acer/Desktop/ml/Test 1/CKDu_processed.csv')

# Extract data and target arrays
data = df.values[:, 1:33] # 32 parameters
target = df.values[:, 0] # group

# Filter data and target to only include first group
group_1_data = data[target == 1]
group_1_target = target[target == 1]

# Create a random forest classifier
rf = RandomForestClassifier()

# Train the classifier on the data for the first group
rf.fit(group_1_data, group_1_target)

# Get feature importances for the first group
importances = rf.feature_importances_

# Create a bar plot of feature importances
features = ['Age', 'S.creatinine', 'Na', 'Mg', 'K', 'Ca', 'Li', 'Be', 'Al', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Ga', 'As', 'Se', 'Rb', 'Sr', 'Ag', 'Cd', 'In', 'Cs', 'Ba', 'Hg', 'Tl', 'Pb', 'Bi', 'U']
plt.bar(features, importances)
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.xlabel('Feature')
plt.title('Feature Importances for Group 1')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# load the data and split into features and targets
CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\CKDu only.csv').values
X = CKDU[:, 1:33]
y = CKDU[:, 0]

# define and fit a random forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# calculate feature importances
importance = rf.feature_importances_

# create a list of feature names
features = ['Age', 'S.creatinine', 'Na', 'Mg', 'K', 'Ca', 'Li', 'Be', 'Al', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Ga', 'As', 'Se', 'Rb', 'Sr', 'Ag', 'Cd', 'In', 'Cs', 'Ba', 'Hg', 'Tl', 'Pb', 'Bi', 'U']

# plot the feature importances
plt.bar(features, importance)
plt.xticks(rotation=90)
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.show()
```
