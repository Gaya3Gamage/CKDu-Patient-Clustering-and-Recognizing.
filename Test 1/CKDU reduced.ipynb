```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np

# Load the dataset
CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\10featuresCKDuOnly.csv')
X = CKDU.iloc[:, 1:13]  # Features
y = CKDU.iloc[:, 0]     # Target

# Reduce the dataset to two features using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a DataFrame with the reduced features
df_pca = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])

# Add the target column to the DataFrame
df_pca['Target'] = y

# Calculate the centers of the 5 target classes
centers = []
for target in range(1, 6):
    center = np.mean(df_pca[df_pca['Target'] == target].iloc[:, :2], axis=0)
    centers.append(center)

# Calculate the unit vectors of the centers
unit_vectors = [center / np.linalg.norm(center) for center in centers]

# Plot the unit vectors in separate colors
colors = ['red', 'green', 'blue', 'orange', 'purple']

plt.figure(figsize=(8, 8))
for i in range(len(unit_vectors)):
    plt.arrow(0, 0, unit_vectors[i][0], unit_vectors[i][1], color=colors[i], head_width=0.1)
    plt.text(unit_vectors[i][0], unit_vectors[i][1], f'Unit Vector {i+1}', color=colors[i], fontsize=12)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Unit Vectors of Cluster Centers')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the dataset
CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\10features.csv')
X = CKDU.iloc[:, 1:13]  # Features
y = CKDU.iloc[:, 0]     # Target

# Perform PCA to reduce the dataset to two features
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a DataFrame for the reduced dataset
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
df_pca['Target'] = y

# Get the centers of the 5 classes
centers = df_pca.groupby('Target').mean()

# Plot the data points
plt.figure(figsize=(10, 6))
colors = ['red', 'green', 'blue', 'orange', 'purple']
for target, color in zip(centers.index, colors):
    subset = df_pca[df_pca['Target'] == target]
    plt.scatter(subset['PC1'], subset['PC2'], color=color, label=f'Target {target}')
    plt.annotate(f'Target {target}', xy=(centers.loc[target, 'PC1'], centers.loc[target, 'PC2']),
                 xytext=(centers.loc[target, 'PC1'] + 0.1, centers.loc[target, 'PC2'] + 0.1),
                 arrowprops=dict(facecolor=color, arrowstyle='->'))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA - Centers of 5 Classes')
plt.legend()
plt.grid(True)
plt.show()

# Calculate unit vectors for each center
unit_vectors = centers.div(centers.pow(2).sum(axis=1).pow(0.5), axis='index')

# Print the unit vectors for each center
print("Unit Vectors for Each Center:")
print(unit_vectors)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\10features normalized new.csv')
data = CKDU.iloc[:, 12:24]  # Features
target = CKDU.iloc[:, 0]

# Apply KMeans clustering and calculate inertia for different number of clusters
max_clusters = 30
inertia_values = []
for n_clusters in range(1, max_clusters + 1):
    model = KMeans(n_clusters=n_clusters, n_init=10, max_iter=3)
    model.fit(data)
    inertia_values.append(model.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_clusters + 1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-Cluster Sum of Squares (Inertia)')
plt.title('Elbow Method for KMeans Clustering')
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from imblearn.over_sampling import RandomOverSampler

# Load the dataset
CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\CKDu_processed.csv')
X = CKDU.iloc[:, 1:13]  # Features
y = CKDU.iloc[:, 0]     # Target

# Apply RandomOverSampler to balance the dataset
oversampler = RandomOverSampler()
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Split the resampled dataset into training and test sets while ensuring equal number of samples in each class
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, stratify=y_resampled)

# Create a Random Forest Classifier
clf = RandomForestClassifier()

# Train the classifier on the training set
clf.fit(X_train, y_train)

# Predict the target values for the test set
y_pred = clf.predict(X_test)
# Print the predicted and actual target values
print("Predicted target:", y_pred)
print("Actual target:", y_test)

# Calculate and print the accuracy
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Count the number of data points in each target class in the training set
target_counts = y_train.value_counts()
print("Number of train data points in each target class:")
print(target_counts)

# Combine the training and test sets
X_combined = pd.concat([X_train, X_test])
y_combined = pd.concat([y_train, y_test])

# Count the number of data points in each target class in the combined dataset
combined_target_counts = y_combined.value_counts()
print("Number of data points in each target class (Combined Dataset):")
print(combined_target_counts)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
from sklearn.metrics import silhouette_score

CKDU = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\10features normalized.csv.csv')
data = CKDU.iloc[:, 1:11]  # Features

# Try different bandwidth values and calculate the silhouette score for each
bandwidth_values = np.linspace(1000, 10000, 20)  # You can adjust the range based on your dataset and requirements
silhouette_scores = []

for bandwidth in bandwidth_values:
    model = MeanShift(bandwidth=bandwidth)
    labels = model.fit_predict(data)
    if len(np.unique(labels)) > 1:  # Ensure there is more than one cluster
        score = silhouette_score(data, labels)
        silhouette_scores.append(score)

# Find the bandwidth with the highest silhouette score
best_bandwidth = bandwidth_values[np.argmax(silhouette_scores)]

print("Optimal Bandwidth:", best_bandwidth)
```
