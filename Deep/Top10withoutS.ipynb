```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint

# Load your dataset
dataset = pd.read_csv(r'C:\Users\acer\Desktop\ml\Deep\Top10CkduOnlyWithoutS.csv')
data = dataset.iloc[:, 1:11]  # Select columns 1 to 32 (32 features)
target = dataset.iloc[:, 0]

# Standardize your data
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(data)

# Assuming 'target' is your array of integer labels
from tensorflow.keras.utils import to_categorical
new_target = to_categorical(target)

# Split your data into training and testing sets
train_data, test_data, train_target, test_target = train_test_split(
    standard_scaled_data, new_target, test_size=0.2)

# Create your neural network model
model = Sequential()
model.add(Dense(512, input_dim=9, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Create a ModelCheckpoint callback to save the best model
checkpoint = ModelCheckpoint(
    'best_model_top10.h5',  # Specify the filename for the best model
    monitor='val_loss',
    save_best_only=True,
    mode='auto'
)

# Train your model and use the checkpoint
history = model.fit(train_data, train_target, epochs=1000,
                    validation_data=(test_data, test_target),
                    callbacks=[checkpoint])

# The best model based on validation loss will be saved as 'best_model.h5'
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Load your dataset
dataset = pd.read_csv(r'C:\Users\acer\Desktop\ml\Test 1\CKDu_processed.csv')

# Assuming the first column contains labels and the rest are features
data = dataset.iloc[:, 1:33]
target = dataset.iloc[:, 0]

# Standardize the input features
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(data)

# Convert target labels to one-hot encoded format
new_target = to_categorical(target)

# Split the data into training and testing sets
train_data, test_data, train_target, test_target = train_test_split(standard_scaled_data, new_target, test_size=0.2)

# Create a Sequential model
model = Sequential()

# Add layers to the model
model.add(Dense(416, input_dim=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(6, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Display the model summary
model.summary()
# Train the model
history = model.fit(train_data, train_target, epochs=1000, validation_data=(test_data, test_target))
```
